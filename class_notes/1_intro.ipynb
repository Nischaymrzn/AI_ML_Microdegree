```python
# Take two numbers as input from the user and perform addition, subtraction, multiplication, 
# and division operations and display the results.

num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
print(f"Sum of {num1} and {num2} is {num1+num2}")
print(f"Sub of {num1} and {num2} is {num1-num2}")
print(f"Multiplication of {num1} and {num2} is {num1*num2}")
print(f"Division of {num1} and {num2} is {num1/num2}")
```

## Conditional Statements

```python
print(num1, num2)
```

```python
if (num1%num2) == 0:
    print(f"{num1} is exactly divisible by {num2}. {int(num2)} * {int(num1/num2)} =  {int(num1)}.")
else:
    print(f"{num1} is not exactly divisible by {num2}.")
```

```python
age = int(input("Enter a person's age: "))

if age<=16:
    print("Person is a child.")
elif age>16 and age<65:
    print("Person is an adult.")
else:
    print("Person is a senior citizen.")
```

```python
# Take two numbers as input from the user and also take an operator as input (+, -, *, /).
# Perform the operation based on the operator provided and display the result.
```

```python
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
operator = input("Enter + if addition, - if substration, * if multiplication, and / if division: ")
if operator == '+':
    print(f'The sum is: {a + b}')
elif operator == '-':
    print(f'The difference is: {a - b}')
elif operator == '*':
    print(f'The product is: {a * b}')
elif operator == '/':
    if b == 0:
        print('Divide by zero error.')
    else:
        print(f'The division result is {a / b}')
else:
    print('Invalid Operator')
```

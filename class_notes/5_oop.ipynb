# Object-Oriented Programming (OOP)

OOP is a programming paradigm that organizes code around "objects" rather than functions and logic. The objects represent real-world entities.

**Pillars of OOP:**
1. Encapsulation: Bundling data and methods together
2. Abstraction: Hiding complex details, showing only the essentials
3. Inheritance: Creating new classes from existing ones
4. Polymorphism: Same declarations/definitions, but different implementation

## Class:

Class is a blueprint for creating objects. It consists of attributes (data) and methods (functions) that the objects will have.

## Object:

An object is an instance of a class. When a class is defined, no memory is allocated until an object is created.

```python
class Student:
    # attributes and methods
    school = "ABC High School"
    pass
```

```python
student1 = Student()
student1.name = "John"
print(student1.name)
print(student1.school)

student2 = Student()
student2.address = "Kathmandu"
print(student2.address)
print(student2.school)

student3 = Student()
student3.age = 20
print(student3.age)
print(student3.school)
```

### Constructor

The constructor is a special method called `__init__` that is automatically executed when an object is created. It initializes the object's attributes

```python
# Person class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")
```

```python
person1 = Person("Ramesh", 25)
print(person1, type(person1))

person1.greet()
```

```python
# Using OOP concepts for bank account management
class BankAccount:
    def __init__(self, account_holder, balance=0):
        # Initialize account holder and balance attributes while creating an object
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        # Add amount to balance
        if amount > 0:
            self.balance = self.balance + amount
            # self.balance += amount
            print(f"Deposited {amount}. New balance is {self.balance}.")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        # Subtract amount from balance if sufficient funds
        if 0 < amount <= self.balance:
            self.balance = self.balance - amount
            # self.balance -= amount
            print(f"Withdrew {amount}. New balance is {self.balance}.")
        else:
            print("Insufficient funds or invalid withdrawal amount.")

    def get_balance(self):
        # Returns the current balance
        print(f"Current balance is {self.balance}.")
```

## Encapsulation and Data Hiding


Encapsulation means bundling data (attributes) and methods together and restricting direct access to some components.

![Public, Private and Protected](https://blogger.googleusercontent.com/img/b/R29vZ2xl/AVvXsEgyror7GsJEoRMVXdV00NBAA1iLENlE-snFJekhYki8eTxgDj-l06TT2NroaqEsuTLJu7SL5MXF6q7aLdVFaIhS_xsMIMBH-IE2wovAT1X_U7w-z-ide39t4L6Zs3RvO5A-7G-X_fD4nu0/s1600/Encapsulation+in+Python+OOP.bmp)

```python
class BankAccount:
    def __init__(self, account_number, amount=0):
        # Initialize account holder and balance attributes while creating an object
        self.account_number = account_number
        self._balance = amount  # Protected attribute
        self.__tpin = "1234"  # Private attribute (name mangling)


    def get_balance(self):
        # Returns the current balance
        return self._balance


    def withdraw(self, amount, pin):
        # Subtract amount from balance if sufficient funds and correct tpin
        if pin == self.__tpin:
            if amount <= self._balance:
                self._balance -= amount
                print(f"Withdrew {amount}. New balance is {self._balance}.")
            else:
                print("Insufficient funds.")
        else:
            print("Invalid TPIN.")

    def __check_tpin(self, pin):
        # Private method to check TPIN
        return pin == self.__tpin
```

### Getters and Setters (@property)

Property decorators provide controlled access to attributes

```python
class BankAccount_Modified:
    def __init__(self, account_number, amount=0):
        self.account_number = account_number
        self._balance = amount  # Protected attribute
        self.__tpin = "12345"  # Private attribute (name mangling)

    @property
    def balance(self):
        # Getter for balance
        return self._balance
    
    @balance.setter
    def balance(self, amount):
        # Setter for balance with validation, allows assigning new amounts if needed
        if amount >= 0:
            self._balance = amount
        else:
            print("Balance cannot be negative.")

    
    @property
    def tpin(self):
        # Getter for tpin
        return "***" # Do not return actual tpin for security reasons
    
    @tpin.setter
    def tpin(self, new_tpin):
        # Setter for tpin with validation
        if len(new_tpin) == 5 and new_tpin.isdigit():
            self.__tpin = new_tpin
        else:
            print("TPIN must be a 5-digit number.")
```

```python
acc1 = BankAccount_Modified("889900111111111", 80000)
acc1.balance #Gets the balance
```

```python
acc1.balance = 90000  #Sets a new balance
acc1.balance
```

```python
acc1.tpin = "54321"  #Sets a new tpin
acc1.tpin
```

## Inheritance

Inheritance allows a class to inherit attributes and methods from another class.

![Types of Inheritance](https://bhrighu.in/vendor/assets/img/blog/Types-of-Inheritance-in-Python.jpg)

```python
class Person:
    """Parent Class (Base Class)"""
    def __init__(self, name, age, nid):
        self.name = name
        self.age = age
        self.nid = nid

    def introduce(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old. My National ID number is {self.nid}.")
    
    def talk(self):
        print("I can talk!")



class Student(Person):
    """Child Class (Derived Class)"""
    def __init__(self, name, age, nid, student_id, major):
        super().__init__(name, age, nid)  # Call the constructor of the parent class
        self.student_id = student_id
        self.major = major

    def study(self):
        print(f"I am studying {self.major}.")
```

```python
p1 = Person("Ramesh", 25, "1234567890")
p1.introduce()
p1.talk()
```

```python
s1.study()
```

```python
s1.introduce()  # Inherited method
s1.talk()       # Inherited method
```

## Polymorphism

Polymorphism means "many forms" - same method name, different implementations.

```python
a = 5
b = 10

c = "10"
d = "20"

fname = "John"
lname = " Doe"


print(a + b)
print(fname + lname)
print(c + d)
```

```python
class Person:
    """Parent Class (Base Class)"""
    def __init__(self, name, age, nid):
        self.name = name
        self.age = age
        self.nid = nid

    def introduce(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old. My National ID number is {self.nid}.")
    
    def talk(self):
        print("I can talk!")



class Student(Person):
    """Child Class (Derived Class)"""
    def __init__(self, name, age, nid, student_id, major):
        super().__init__(name, age, nid)  # Call the constructor of the parent class
        self.student_id = student_id
        self.major = major

    def study(self):
        print(f"I am studying {self.major}.")

    
    def introduce(self):
        print(f"Hello, my name is {self.name}, I am {self.age} years old, my National ID number is {self.nid}, and I am majoring in {self.major}.")
```

```python
p2 = Person("Ram", 28, "1122334455")
p2.introduce()
p2.talk()
```

```python
s2 = Student("Gita", 21, "6677889900", "S54321", "Mathematics")
s2.introduce()  # Polymorphic method
s2.talk()       # Inherited method
s2.study()      # Student class method
```

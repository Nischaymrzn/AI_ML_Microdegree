## For Loop

for loop is based on the number of times a specific block of code / statements are to be executed.

```python
students = ["ST100", "ST105", "ST110", "ST115", "ST120", "ST125"]

check = input("Enter student ID to check: ")

for i in students:
    if i == check:
        print(f"{check} is enrolled in the course.")
        break
else:
    print(f"{check} is not enrolled in the course.")
```

```python
for i in range(5):
    print(i)
```

```python
for i in range(5, 15, 2):
    print(i)
```

```python
for i in range(1, 3):
    print(f"i = {i}")
```

```python
for j in range(10, 15):
    print(f"j = {j}")
```

```python
# nested loop
for i in range(1, 3):
    for j in range(10, 15):
        print(f"i = {i}, j = {j}")
```

# While Loop

While loop is based on the condition to run a specific block of statements

```python
count = 0
print(f"Initial count value: {count}")
print("Starting while loop...")
while count <= 5:
    print(f"Current count value: {count}")
    count = count + 1
print(f"While loop ended with final count value of {count}.")
```

```python
from random import randint

number_to_guess = randint(1, 10)

while True:
    guess = int(input("Guess a number between 1 and 10: "))

    if guess < number_to_guess:
        print("Too low! Try again.")
    elif guess > number_to_guess:
        print("Too high! Try again.")
    else:
        print(f"Congratulations! You've guessed the correct number. {number_to_guess} is the right answer.")
        break


print("Game Over.")
```

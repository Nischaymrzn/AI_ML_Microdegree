### Assignment 1
#### Write a Python program that takes a user’s name, age, and three subject marks.
- Display the name and age.
- Calculate and print the total and average marks.
- Print whether the student has passed or failed (pass mark = 40 in all subjects).

```python
name = input("Enter your name: ")
age = int(input("Enter your age: "))

math = float(input("Enter marks for Math: "))
science = float(input("Enter marks for Science: "))
english = float(input("Enter marks for English: "))

print("Name:", name)
print("Age:", age)

total = math + science + english
average = total / 3

print("Total Marks:", total)
print("Average Marks:", average)

if math >= 40 and science >= 40 and english >= 40:
    print("Result: Passed")
else:
    print("Result: Failed")
```

### Assignment 2
#### Write a Python program that:
- Takes 10 numbers as input from the user and stores them in a list.
- Prints all even numbers and their sum.
- Finds and prints the largest and smallest numbers in the list.

```python
numbers = []

for i in range(10):
    num = int(input(f"Enter number {i+1}: "))
    numbers.append(num)

print("Even numbers are:")
sum = 0
for n in numbers:
    if n % 2 == 0:
        print(n)
        sum = sum + n

print("Sum of even numbers:", sum)

largest = numbers[0]
smallest = numbers[0]

for n in numbers:
    if n > largest:
        largest = n
    if n < smallest:
        smallest = n

print("Largest number:", largest)
print("Smallest number:", smallest)
```

### Assignment 3
#### Write a Python program that:
- Creates a dictionary with names of 5 students as keys and their marks as values.
- Prints the names of students who scored more than 80.

```python
students = {
    "Nischay": 85,
    "Khushal": 81,
    "Trishan": 90,
    "Shreya": 67,
    "Rabin": 82
}

print("Students scoring more than 80:")

for name, marks in students.items():
    if marks > 80:
        print(name)
```

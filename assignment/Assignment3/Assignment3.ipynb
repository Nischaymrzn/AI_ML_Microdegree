### Question 1: Library Book Lending Analysis

You have a CSV file `library_data.csv` with the following columns: `Book_Title`, `Genre`, `Days_Borrowed`, `Member_Type`.

**Tasks:**
1. Read the CSV file into a pandas DataFrame
2. Group the data by `Genre` and calculate the total days borrowed and average days borrowed for each genre
3. Find which genre has the highest total borrowing days
4. Export the grouped results to a new CSV file called `genre_summary.csv`

**Sample Data to Create:**
```
Book_Title,Genre,Days_Borrowed,Member_Type
Python Basics,Technology,14,Student
Mystery Novel,Fiction,7,Adult
Data Science,Technology,21,Student
Fantasy Quest,Fiction,10,Adult
AI Guide,Technology,14,Senior
Crime Story,Fiction,5,Adult
```

````python
import pandas as pd

data = {
    "Book_Title": ["Python Basics", "Mystery Novel", "Data Science", "Fantasy Quest", "AI Guide", "Crime Story"],
    "Genre": ["Technology", "Fiction", "Technology", "Fiction", "Technology", "Fiction"],
    "Days_Borrowed": [14, 7, 21, 10, 14, 5],
    "Member_Type": ["Student", "Adult", "Student", "Adult", "Senior", "Adult"]
}

df = pd.DataFrame(data)
df.to_csv("library_data.csv", index=False)

df = pd.read_csv("library_data.csv")

genre_group = df.groupby("Genre")["Days_Borrowed"].agg(
    Total_Days="sum",
    Average_Days="mean"
)

highest_genre = genre_group.sort_values("Total_Days", ascending=False).index[0]

print("Genre with highest total borrowing days:", highest_genre)

genre_group.to_csv("genre_summary.csv")
````

### Question 2: Student Performance Analysis with Pivot Table

You have student exam data with columns: `Student_Name`, `Subject`, `Score`, `Semester`.

**Tasks:**
1. Create a DataFrame with at least 12 rows of student data (3 students, 2 subjects, 2 semesters)
2. Use `pivot_table()` to create a table showing average scores for each student across different subjects
3. Use `crosstab()` to find the count of exams taken by each student in each semester
4. Save the pivot table to an Excel file called `student_performance.xlsx`

**Hint:** Your data should look something like:
- Students: Alice, Bob, Charlie
- Subjects: Math, Science
- Semesters: Fall, Spring

````python
data = {
    "Student_Name": ["Alice", "Alice", "Alice", "Alice",
                     "Bob", "Bob", "Bob", "Bob",
                     "Charlie", "Charlie", "Charlie", "Charlie"],
    "Subject": ["Math", "Science", "Math", "Science"] * 3,
    "Score": [85, 90, 88, 92, 70, 75, 72, 78, 95, 90, 93, 97],
    "Semester": ["Fall", "Fall", "Spring", "Spring"] * 3
}

df = pd.DataFrame(data)
df.to_csv("students.csv", index=False)

pivot = pd.pivot_table(df, values="Score", index="Student_Name", columns="Subject", aggfunc="mean")

exam_count = pd.crosstab(df["Student_Name"], df["Semester"])

with pd.ExcelWriter("student_performance.xlsx") as writer:
    pivot.to_excel(writer, sheet_name="Average Scores")
    exam_count.to_excel(writer, sheet_name="Exam Count")
````

### Question 3: Employee Data Merging

You have two CSV files:
- `employees.csv` with columns: `Employee_ID`, `Name`, `Department`
- `salaries.csv` with columns: `Employee_ID`, `Salary`, `Bonus`

**Tasks:**
1. Create both CSV files with at least 5 employees each
2. Read both files into separate DataFrames
3. Merge the two DataFrames using `Employee_ID` as the key
4. Calculate the total compensation (Salary + Bonus) for each employee
5. Group by `Department` and find the average total compensation per department
6. Export the final result to `department_compensation.csv`

````python
employees = {
    "Employee_ID": [1, 2, 3, 4, 5],
    "Name": ["Ram", "Sita", "Hari", "Gita", "John"],
    "Department": ["IT", "HR", "IT", "Finance", "HR"]
}

salaries = {
    "Employee_ID": [1, 2, 3, 4, 5],
    "Salary": [50000, 45000, 55000, 60000, 48000],
    "Bonus": [5000, 3000, 7000, 8000, 4000]
}

pd.DataFrame(employees).to_csv("employees.csv", index=False)
pd.DataFrame(salaries).to_csv("salaries.csv", index=False)

emp_df = pd.read_csv("employees.csv")
sal_df = pd.read_csv("salaries.csv")

merged = pd.merge(emp_df, sal_df, on="Employee_ID")

merged["Total_Compensation"] = merged["Salary"] + merged["Bonus"]
print(merged)

dept_summary = merged.groupby("Department")["Total_Compensation"].mean()

dept_summary.to_csv("department_compensation.csv")
````

### Question 4: Product Inventory Concatenation

You manage inventory across 3 warehouses, each with their own data file.

**Tasks:**
1. Create 3 separate DataFrames representing inventory from Warehouse A, B, and C
   - Columns: `Product_ID`, `Product_Name`, `Stock`, `Warehouse`
   - Each warehouse should have 4-5 products
2. Concatenate all three DataFrames into a single inventory DataFrame
3. Group by `Product_Name` and calculate the total stock across all warehouses
4. Use `agg()` to find both the sum and count of stock for each product
5. Save the complete inventory to `complete_inventory.csv`

````python
wa = pd.DataFrame({
    "Product_ID": [1,2,3,4],
    "Product_Name": ["Pen", "Notebook", "Marker", "Eraser"],
    "Stock": [100, 50, 60, 40],
    "Warehouse": "A"
})

wb = pd.DataFrame({
    "Product_ID": [1,2,3,5],
    "Product_Name": ["Pen", "Notebook", "Marker", "Bag"],
    "Stock": [90, 40, 70, 20],
    "Warehouse": "B"
})

wc = pd.DataFrame({
    "Product_ID": [1,3,4,5],
    "Product_Name": ["Pen", "Marker", "Eraser", "Bag"],
    "Stock": [110, 65, 55, 30],
    "Warehouse": "C"
})

inventory = pd.concat([wa, wb, wc], ignore_index=True)

summary = inventory.groupby("Product_Name").agg(
    Total_Stock=("Stock", "sum"),
    Count_Entries=("Stock", "count")
)
print(summary)


inventory.to_csv("complete_inventory.csv", index=False)
````

### Question 5: Customer Orders Analysis

You have customer order data with columns: `Order_ID`, `Customer_ID`, `Product_Category`, `Order_Amount`, `Order_Date`.

**Tasks:**
1. Create a DataFrame with at least 15 orders from 5 different customers
2. Read the data from a CSV file (create it first)
3. Group by `Customer_ID` and use `agg()` to calculate:
   - Total order amount
   - Average order amount
   - Number of orders
4. Create a pivot table showing total order amounts by `Customer_ID` and `Product_Category`
5. Find the top 3 customers by total order amount
6. Export the customer summary to an Excel file with two sheets:
   - Sheet 1: Customer aggregation results
   - Sheet 2: Pivot table results

````python
data = {
    "Order_ID": range(1,16),
    "Customer_ID": ["C1","C2","C3","C4","C5"] * 3,
    "Product_Category": ["Electronics","Clothing","Food"] * 5,
    "Order_Amount": [1200,500,200,1500,300,250,800,1000,400,1100,600,150,900,750,350],
    "Order_Date": pd.date_range("2024-01-01", periods=15)
}

pd.DataFrame(data).to_csv("orders.csv", index=False)

df = pd.read_csv("orders.csv")

customer_summary = df.groupby("Customer_ID").agg(
    Total_Amount=("Order_Amount", "sum"),
    Average_Amount=("Order_Amount", "mean"),
    Order_Count=("Order_ID", "count")
)

pivot = pd.pivot_table(df, values="Order_Amount", index="Customer_ID", columns="Product_Category", aggfunc="sum")

top3 = customer_summary.sort_values("Total_Amount", ascending=False).head(3)

print(top3)

with pd.ExcelWriter("customer_orders_summary.xlsx") as writer:
    customer_summary.to_excel(writer, sheet_name="Customer Summary")
    pivot.to_excel(writer, sheet_name="Pivot Table")
````
